```python
# TechCrush AI Bootcamp - Week 6 Deep Learning Practical
# Project: Handwritten Digit Classification with MNIST

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
import matplotlib.pyplot as plt

print("--- Step 1: Loading and Preprocessing Data ---")
# 1. Import the MNIST dataset directly from keras.datasets
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# 2. Preprocess the images: scale pixel values to be between 0 and 1
# Neural networks learn much faster when numbers are normalized.
x_train, x_test = x_train / 255.0, x_test / 255.0

print(f"Successfully loaded {len(x_train)} training images and {len(x_test)} testing images.")

print("\n--- Step 2: Building the Neural Network Architecture ---")
# 3. Build a Sequential model in Keras
model = Sequential([
    # Input Layer: Flattens the 28x28 pixel 2D image into a 1D array of 784 pixels
    Flatten(input_shape=(28, 28)),

    # Hidden Layer: 128 neurons doing the heavy lifting with 'relu' activation
    Dense(128, activation='relu'),

    # Regularization: Randomly turns off 20% of neurons to prevent overfitting
    Dropout(0.2),

    # 4. Final Output Layer: 10 categories (digits 0-9) using 'softmax' activation
    Dense(10, activation='softmax')
])
print("Neural Network assembled successfully.")

print("\n--- Step 3: Compiling the Model ---")
# 5. Compile the model by defining the optimizer and loss function
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
print("Model compiled.")

print("\n--- Step 4: Training the AI ---")
# 6. Train the model on the data for 5 epochs (full passes through the dataset)
model.fit(x_train, y_train, epochs=5)

print("\n--- Step 5: Final Evaluation ---")
# 7. Evaluate how well the AI performs on the unseen testing data
test_loss, test_acc = model.evaluate(x_test,  y_test, verbose=2)
print(f"\n✅ Final Model Accuracy on unseen data: {test_acc*100:.2f}%")
```
